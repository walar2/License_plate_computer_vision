```python
from ultralytics import YOLO

# change this if your yaml is somewhere else under /content
DATA_YAML = "/content/data.yaml"

model = YOLO("yolov8n.pt")  # or yolov8s.pt / m / l

model.train(
    data=DATA_YAML,   # data.yaml with train/val/test paths
    epochs=50,
    imgsz=640,
    batch=16,
    workers=2,
    project="yolov8_runs",
    name="exp_custom",
)
```
